Now let's extract the data in the csv file using the csv module in python

```python
import csv

file = open('cereal.csv')
type(file)
csv_file_reader_object = csv.reader(file)

# first we extract the column headings
headings = []
headings = next(csv_file_reader_object)
print(headings)
print('-------------')

# now for the data 
rows = []
for row in csv_file_reader_object:
    rows.append(row)

print("Number of Rows : " , len(rows))
print("Rows:")
print(rows)
```

Data extraction successful !

Numpy is a very important, useful and famous library as it offers multi-dimensional array capabilities. Lists store values, numpy arrays have built in functions that we can use to manipulate the data en masse. 

Now let us turn our attention to using Numpy to manipulate the data. Remember that 'rows' contains our data and 'columns' contains the headings.

```python
import numpy as np 

# The data is in python lists. To use numpy we need to convert them into numpy arrays
# To keep it simple for analysis lets use just some of this data

rows_array = np.array(rows)

print("The array - \n", rows_array)
print("Dimension- \n", rows_array.ndim) #prints how many rows present
print("Shape - \n" ,rows_array.shape) #prints the shape of the array in (rows, columns)
```

```python
# now lets see some of the arithmetic operations offered by numpy
# first we will need the numerical columns in the dataset. Lets use calories and protein

calories = rows_array[:,3] # column slicing
protein = rows_array[:,4] # negative slicing

calories = calories.astype(np.int64)
protein = protein.astype(np.int64)

print("Calories array - \n", calories)
print("Protein array - \n", protein)
```

```python
# lets find out more about the calories array
print("Datatype of the calories array-", calories.dtype) #prints the datatype of the array

print("Net array size -", calories.size) #prints net size of the array.
```

```python
# how about if we want a particular element? 

print(rows_array[1,0])
```

```python
# now for operations. First lets square all elements in the second column 
# and find the sine value of elements in the third column

calorie_squares = np.square(calories)
protein_sines = np.sin(protein)
calorie_squares_alternate = np.power(calories,2)

print(" Calories ^2 - \n", calorie_squares)
print(" sin(protein) - \n", protein_sines)
print("Calories^2 alternate-\n", calorie_squares_alternate)
```

```python
# lets see how the add, subtract and multiply commands work

calories_plus_protein = np.add(calories,protein)
calories_minus_protein = np.subtract(calories, protein)
calories_times_calories = np.multiply(calories, calories)

print("addition -\n ", calories_plus_protein)
print("subtraction -\n ", calories_minus_protein)
print("multiplication -\n ", calories_times_calories) # compare this with calories^2 in the earlier cell
```

Lets try stacking 

```python
# For more clarity on this lets use an even smaller set of values from the 
# calories and protein values

a = calories[0:3]
b = protein[0:3]

print(a)
print(b)
```

```python
horizontal = np.hstack((a,b))
vertical = np.vstack((a,b))
column = np.column_stack((a,b))

print("horizontal stacking - \n", horizontal)
print("vertical stacking - \n",vertical)
print("stacking columns - \n",column)
```

Shape manipulation is an important part of numpy usage. Lets see how we can do that:

```python
# lets reshape the vertical stack
print("Vertical Shape : " , vertical.shape)
print("unravelled -\n", vertical.ravel())

# reshape command-
reshaped = vertical.reshape(3,2)
print("after using reshape-\n", reshaped)
print("original array-\n", vertical)

#resize command:
vertical.resize(3,2)
print("original array has changed-\n", vertical)
```

An important difference between reshape and resize is highlighted here. Reshape return a new array leaving the old one unchanged while resize changes the original array.
